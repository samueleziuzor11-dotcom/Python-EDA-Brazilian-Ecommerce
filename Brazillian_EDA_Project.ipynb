# 8. Recreate top_states AFTER mapping so it uses full names

```python
top_states = df['customer_state'].value_counts().head(10)

plt.figure(figsize=(14,8))
sns.barplot(x=top_states.index, y=top_states.values)
plt.title('Top 10 States by Number of Customers')
plt.xlabel('State')
plt.ylabel('Number of Customers')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

## Insight: Geographic Concentration Risk

São Paulo alone accounts for roughly 42% of all customers.
The top 3 states (SP, RJ, MG) together represent over 60% of the entire customer base.

Analytical takeaway: This level of concentration means the business is
highly exposed to regional risk. Any logistics disruption, economic downturn,
or competitor entering São Paulo directly threatens the majority of revenue.
A data-driven growth strategy would prioritize customer acquisition in
underrepresented states like the North and Northeast regions.

# 9. Top 10 Cities by Customer Count

```python
top_cities = df['customer_city'].value_counts().head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=top_cities.values, y=top_cities.index, palette='viridis')
plt.title('Top 10 Cities by Number of Customers', fontsize=16)
plt.xlabel('Number of Customers')
plt.ylabel('City')
plt.tight_layout()
plt.show()
```

## Insight: City-Level Dominance

São Paulo city alone has more customers than the next 4 cities combined.
Rio de Janeiro is a distant second, followed by Belo Horizonte.

Analytical takeaway: The extreme dominance of São Paulo city suggests
that e-commerce adoption in Brazil is still heavily urbanized and concentrated
in one metro area. This presents a clear opportunity  mid-sized cities like
Curitiba and Porto Alegre show meaningful customer bases and could be
high-ROI targets for focused marketing campaigns with lower competition
than São Paulo.

# 10. # Map states to regions

```python
region_map = {
    'São Paulo': 'Southeast', 'Rio de Janeiro': 'Southeast',
    'Minas Gerais': 'Southeast', 'Espírito Santo': 'Southeast',
    'Paraná': 'South', 'Santa Catarina': 'South',
    'Rio Grande do Sul': 'South',
    'Bahia': 'Northeast', 'Ceará': 'Northeast', 'Pernambuco': 'Northeast',
    'Maranhão': 'Northeast', 'Paraíba': 'Northeast', 'Rio Grande do Norte': 'Northeast',
    'Alagoas': 'Northeast', 'Sergipe': 'Northeast', 'Piauí': 'Northeast',
    'Mato Grosso': 'Central-West', 'Mato Grosso do Sul': 'Central-West',
    'Goiás': 'Central-West', 'Distrito Federal': 'Central-West',
    'Amazonas': 'North', 'Pará': 'North', 'Rondônia': 'North',
    'Roraima': 'North', 'Amapá': 'North', 'Acre': 'North', 'Tocantins': 'North'
}

df['region'] = df['customer_state'].map(region_map)

region_counts = df['region'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(region_counts.values, labels=region_counts.index,
        autopct='%1.1f%%', colors=sns.color_palette('viridis', len(region_counts)))
plt.title('Customer Distribution by Region', fontsize=16)
plt.tight_layout()
plt.show()
```

## Insight: Regional Inequality in E-Commerce Reach

The Southeast region captures over 60% of all customers despite Brazil
having 5 distinct regions with large populations elsewhere.

Analytical takeaway: This is not just a business observation it reflects
a deeper infrastructure and digital divide in Brazil. The North region,
which covers 45% of Brazil's land area, contributes less than 5% of customers.

From a data science perspective, this raises important questions worth
investigating with additional data:
- Are low numbers in the North due to lack of internet access?
- Are delivery costs making e-commerce impractical there?
- Is there untapped demand that targeted campaigns could unlock?

Identifying *why* a pattern exists is just as important as identifying
that it exists this is what separates exploratory analysis from
actionable insight.
